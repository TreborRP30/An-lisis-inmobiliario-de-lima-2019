📊 **Análisis Estadístico Descriptivo y Detección de Outliers**

En esta fase exploratoria, generamos un resumen estadístico descriptivo utilizando el método .describe() sobre las variables numéricas más críticas para el análisis (Precio, Area_constr_m2, Nro_banios). El objetivo principal de esta operación es auditar la integridad matemática de los datos para comprender su distribución inicial e identificar valores atípicos (outliers) antes de iniciar la etapa de modelado.

Hallazgos clave del diagnóstico:

Identificación de Extremos (Mínimos y Máximos): El análisis de los indicadores min y max expuso anomalías severas que comprometen la calidad del dataset. Como evidencia, la variable Area_constr_m2 arrojó valores mínimos físicamente imposibles para una vivienda (1.0 m²) y máximos desproporcionados (8,000.0 m²). Esto confirma la presencia de ruido originado por errores de digitación al momento de publicar los anuncios web.

Evaluación de la Dispersión:

Al comparar la media (mean) con la mediana (el percentil 50%), se observan distorsiones provocadas por estos valores atípicos. Las propiedades con precios o áreas absurdamente altas están "jalando" el promedio hacia arriba, alejándolo del comportamiento real del mercado (representado entre los cuartiles 25% y 75%).

```python
#Definimos las variables numéricas más críticas de tu dataset
columnas_clave = ['precio_usd', 'area_m2', 'dormitorios', 'banos','precio_m2']

#Generamos un resumen estadístico, mostrando los valores mínimos y máximos
df[columnas_clave].describe().round(2)
```

```python
# 1. Filtro Estadístico (IQR) para variables continuas
columnas_iqr = ['precio_usd', 'area_m2', 'precio_m2']
mascara_iqr = pd.Series(True, index=df.index)

for col in columnas_iqr:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    mascara_iqr = mascara_iqr & (df[col] >= limite_inferior) & (df[col] <= limite_superior)

# 2. Filtro por Reglas de Negocio para dormitorios y baños
# Así eliminamos ese outlier de 19 cuartos y 20 baños, pero conservamos propiedades grandes reales
mascara_logica = (
    (df['dormitorios'] >= 1) & (df['dormitorios'] <= 6) &
    (df['banos'] >= 1) & (df['banos'] <= 6)
)

# 3. Combinamos ambas condiciones en un nuevo DataFrame limpio
df_limpio = df[mascara_iqr & mascara_logica]

# 4. Evaluamos el impacto del proceso de limpieza
filas_eliminadas = df.shape[0] - df_limpio.shape[0]
print(f"📋 Filas iniciales: {df.shape[0]}")
print(f"✨ Filas finales tras limpieza total: {df_limpio.shape[0]}")
print(f"🚫 Total de outliers detectados y eliminados: {filas_eliminadas}")
```

📉 **Análisis de Varianza y Reducción de Dimensionalidad**

En esta fase de limpieza, el enfoque pasó de las variables numéricas a las variables categóricas o booleanas (características binarias como la presencia de Sauna, Sala_de_cine, etc.). En este tipo de datos no se buscan "valores extremos", sino que se audita el desbalance de clases o la baja varianza.

Acciones técnicas realizadas:

Auditoría Automática de Varianza: Se implementó un algoritmo iterativo para escanear todas las columnas del dataset e identificar aquellas que presentaban un comportamiento "casi constante". Se estableció un umbral estadístico estricto: si un único valor dominaba más del 95% de los registros en una columna (por ejemplo, si el 98% de los inmuebles "No" tiene Sala_de_cine), la variable fue marcada para revisión.

Depuración por Carencia de Valor Predictivo: Las características identificadas con nula diversidad fueron eliminadas estratégicamente del DataFrame. ¿Por qué se hizo? En el modelado de datos y Machine Learning, si una variable no fluctúa, no aporta información útil para que el algoritmo encuentre patrones.

Mantener columnas casi idénticas en todas las filas solo aumenta el ruido, consume memoria innecesaria y provoca la "maldición de la dimensionalidad" sin ayudar a predecir el precio del inmueble.

```python
#Auditamos de Baja Varianza (Desbalance)
print("🔍 Analizando columnas casi constantes (más del 95% iguales):\n")

columnas_baja_varianza = []

#Iteramos por todas las columnas del dataset
for col in df.columns:
    # Aseguramos que la columna tenga datos
    if df[col].count() > 0:
        # Calculamos el porcentaje del valor que más se repite
        freq_maxima = df[col].value_counts(normalize=True).iloc[0] * 100

        #Si un solo valor domina más del 95% de la columna...
        if freq_maxima > 95:
            print(f"🚨 {col}: {freq_maxima:.2f}% de los datos son idénticos.")
            columnas_baja_varianza.append(col)

print(f"\nTotal de variables que no aportan valor predictivo: {len(columnas_baja_varianza)}")

# 2. Eliminación de las columnas sin varianza
df.drop(columns=columnas_baja_varianza, inplace=True, errors='ignore')

print(f"Limpieza completada. Ahora te quedan {df.shape[1]} columnas y {df.shape[0]} filas útiles en tu dataset.")
```

📋 **Auditoría Final de Calidad de Datos (Quality Assurance)**

Antes de dar por concluida la fase de limpieza estructural, se ejecutó un control de calidad (QA Check) de tres niveles para garantizar la viabilidad del dataset frente a algoritmos de Machine Learning. Esta auditoría sistemática es un estándar en la industria para prevenir errores silenciosos en la fase de modelado.

Controles ejecutados:

Verificación de Registros Duplicados: Se escaneó el dataset en busca de filas idénticas. En proyectos de web scraping, es habitual extraer el mismo anuncio inmobiliario múltiples veces debido a la paginación dinámica de los sitios web. Eliminar estos duplicados evita el sobreajuste (overfitting) y asegura que las métricas de promedio no estén sesgadas.

Auditoría de Valores Nulos Residuales (NaN): Se realizó un mapeo de las columnas clave para asegurar que, tras las transformaciones y eliminaciones previas, no quedaran campos vacíos ocultos que pudieran interrumpir la ejecución matemática de futuros modelos.

Validación de Tipos de Datos (Dtypes): Se revisó la estructura interna de las variables predictoras (Precio, Area, Dormitorios, etc.). El objetivo de este paso crítico es asegurar que el motor de Pandas reconozca las variables matemáticas como formatos numéricos (float64 o int64) y no como cadenas de texto (object). Ignorar este paso resultaría en fallos de compilación al intentar calcular distancias estadísticas o coeficientes de regresión.

```python
print("--- 1. VERIFICACIÓN DE DUPLICADOS ---")
duplicados = df.duplicated().sum()
print(f"Tienes {duplicados} filas exactamente iguales.")
# Si hay duplicados, los eliminamos
if duplicados > 0:
    df.drop_duplicates(inplace=True)
    print("¡Duplicados eliminados!")

print("\n--- 2. VERIFICACIÓN DE NULOS (TOP 10) ---")
# Esto te mostrará las 10 columnas con más datos faltantes
nulos = df.isnull().sum().sort_values(ascending=False).head(10)
print(nulos[nulos > 0]) # Solo muestra si hay mayores a 0

print("\n--- 3. VERIFICACIÓN DE TIPOS DE DATOS (Dtypes) ---")
# Verificamos si Pandas reconoció bien los números
print("Revisa si alguna de estas numéricas aparece como 'object':")
# Muestra el tipo de dato de nuestras columnas clave
print(df[['Precio', 'Area_constr_m2', 'Nro_banios', 'Dormitorios']].dtypes)
```

🛠️ **Corrección de Tipos de Datos (Casting) y Formateo Numérico**

Durante la auditoría final de calidad de datos (Dtypes), se detectó que la variable clave Dormitorios estaba catalogada como object (cadena de texto) en lugar de un formato numérico.

Esto suele ser provocado por caracteres alfanuméricos residuales del scraping web.

Para solucionarlo, se aplicó una técnica de coerción numérica mediante la función pd.to_numeric(errors='coerce'). Este método fuerza la transformación de la columna a formato de coma flotante (float64), convirtiendo automáticamente cualquier cadena de texto no parseable en valores nulos (NaN), los cuales fueron posteriormente removidos. Esto garantiza la compatibilidad total del dataset con futuros algoritmos de Machine Learning.

```python
#Vemos qué valores únicos raros hay antes de la limpieza (para entender el error)
print("Algunos valores actuales en la columna:", df['Dormitorios'].unique()[:10])

#Forzamos la conversión a número.
#errors='coerce' convierte cualquier texto rebelde (como "Sin dato") en nulo (NaN)
df['Dormitorios'] = pd.to_numeric(df['Dormitorios'], errors='coerce')

#Verificamos si la conversión generó valores nulos y los eliminamos
nulos_generados = df['Dormitorios'].isnull().sum()
if nulos_generados > 0:
    print(f"\nSe encontraron {nulos_generados} textos que no eran números. Eliminando esas filas...")
    df.dropna(subset=['Dormitorios'], inplace=True)

#Verificación final del tipo de dato
print("\n--- NUEVOS TIPOS DE DATOS ---")
print(df[['Precio', 'Area_constr_m2', 'Nro_banios', 'Dormitorios']].dtypes)

#Imprimimos el dataset ya limpio y listo para su análisis
print(f"Limpieza completada. Ahora te quedan {df.shape[1]} columnas y {df.shape[0]} filas útiles en tu dataset.")
```
