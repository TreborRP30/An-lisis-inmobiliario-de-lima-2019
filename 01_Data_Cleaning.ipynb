🧹 **Tratamiento Híbrido de Outliers: Combinando Estadística y Reglas de Negocio**


Durante la fase de limpieza (ETL) en Python de mi proyecto de Análisis Inmobiliario en Lima, me topé con un desafío clásico: la presencia de valores atípicos (outliers). Al explorar los datos, identifiqué registros con precios absurdos de $1 o propiedades listadas con hasta 19 habitaciones, errores que sesgarían por completo cualquier cálculo de rentabilidad y ROI.


Para limpiar el dataset sin correr el riesgo de eliminar propiedades legítimas, decidí diseñar e implementar un enfoque híbrido utilizando máscaras booleanas en Pandas:


📈 1. Filtro Estadístico (Rango Intercuartílico - IQR)
Para las variables financieras y de escala continua (precio_usd, area_m2 y precio_m2), implementé el método matemático IQR. Mediante un bucle for, el algoritmo evalúa estas métricas y recorta los extremos de la distribución. Esto limpia automáticamente errores de digitación en precios y descarta terrenos de escala industrial que no aportan a mi análisis residencial.


🧠 2. Filtro por Reglas de Negocio
Aplicar estadística pura a variables discretas (como el número de dormitorios o banos) habría sido un error, ya que el IQR podría considerar una casa legítima de 5 habitaciones como una "anomalía". Para evitar esta pérdida de datos valiosos, apliqué reglas de negocio basadas en la lógica del mercado, limitando estos espacios a un rango residencial realista (de 1 a 6).


🛠️ El Resultado
Al interceptar ambas máscaras lógicas, logré depurar el dataset con precisión. El resultado es una base de datos robusta, íntegra y libre de ruido. Ahora los datos están listos para ser modelados en SQL y, finalmente, dar vida a mi dashboard interactivo desarrollado netamente con librerías de Python.

```python
#Definimos las variables numéricas más críticas de tu dataset
columnas_clave = ['precio_usd', 'area_m2', 'dormitorios', 'banos','precio_m2']

#Generamos un resumen estadístico, mostrando los valores mínimos y máximos
df[columnas_clave].describe().round(2)
```

```python
#Filtro Estadístico (IQR) para variables continuas
columnas_iqr = ['precio_usd', 'area_m2', 'precio_m2']
mascara_iqr = pd.Series(True, index=df.index)

for col in columnas_iqr:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    mascara_iqr = mascara_iqr & (df[col] >= limite_inferior) & (df[col] <= limite_superior)

#Filtro por Reglas de Negocio para dormitorios y baños
#Así eliminamos ese outlier de 19 cuartos y 20 baños, pero conservamos propiedades grandes reales
mascara_logica = (
    (df['dormitorios'] >= 1) & (df['dormitorios'] <= 6) &
    (df['banos'] >= 1) & (df['banos'] <= 6)
)

#Combinamos ambas condiciones en un nuevo DataFrame limpio
df_limpio = df[mascara_iqr & mascara_logica]

#Evaluamos el impacto del proceso de limpieza
filas_eliminadas = df.shape[0] - df_limpio.shape[0]
print(f"📋 Filas iniciales: {df.shape[0]}")
print(f"✨ Filas finales tras limpieza total: {df_limpio.shape[0]}")
print(f"🚫 Total de outliers detectados y eliminados: {filas_eliminadas}")
```
